```python
#################################
# 1. Setup, Downloads & Imports
#################################
# Download and unzip GloVe embeddings (run this cell separately or at the beginning)
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip -o glove.6B.zip

# Install dependencies
!pip install seaborn ipywidgets

# Import necessary libraries
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader, random_split
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, precision_score, recall_score, f1_score
from collections import Counter
from copy import deepcopy
from tqdm import tqdm
import string
import torch.nn.functional as F
import seaborn as sns
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, HTML
import matplotlib.colors as mcolors
import random

# Set deterministic behavior for reproducibility
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed_all(seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False


#################################
# 2. Data Loading & Preprocessing
#################################
# --- Data Loading Functions ---
def load_data(filepath='Sarcasm_Headlines_Dataset.json'):
    """Load the headlines and their labels from a JSON file."""
    df = pd.read_json(filepath, lines=True)
    return df['headline'], df['is_sarcastic']

def split_data(texts, labels, test_size=0.2, random_state=42):
    """Split the data into training and testing sets."""
    return train_test_split(texts, labels, test_size=test_size, random_state=random_state)

# --- Text Tokenization and Vocabulary ---
def tokenize(text):
    """
    Tokenize a given text by converting it to lowercase, removing punctuation,
    and splitting on whitespace.
    """
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text.split()

def build_vocab(texts):
    """
    Build a vocabulary from the training texts. Reserve index 0 for <PAD> and 1 for <UNK>.
    """
    all_tokens = []
    for text in texts:
        all_tokens.extend(tokenize(text))
    counter = Counter(all_tokens)
    vocab = {token: idx + 2 for idx, (token, _) in enumerate(counter.most_common())}
    vocab["<PAD>"] = 0
    vocab["<UNK>"] = 1
    return vocab

# --- Sequence Conversion & Padding ---
def text_to_sequence(text, vocab):
    """Convert text to a list of token indices using the vocabulary."""
    return [vocab.get(token, vocab["<UNK>"]) for token in tokenize(text)]

def pad_sequence(seq, max_length, pad_value=0):
    """Pad or truncate a sequence to a fixed max length."""
    if len(seq) < max_length:
        return seq + [pad_value] * (max_length - len(seq))
    return seq[:max_length]

def preprocess_texts(texts, vocab, max_length=20):
    """Convert a list of texts to padded sequences."""
    return [pad_sequence(text_to_sequence(text, vocab), max_length) for text in texts]

# --- DataLoader Creation ---
def create_dataloaders(X_train_seq, X_test_seq, y_train, y_test, batch_size=32, val_split=0.1):
    """
    Create DataLoader objects for training, validation, and testing.
    Validation split is taken from the training data.
    """
    X_train_tensor = torch.tensor(X_train_seq, dtype=torch.long)
    X_test_tensor  = torch.tensor(X_test_seq, dtype=torch.long)
    y_train_tensor = torch.tensor(y_train.values, dtype=torch.float32).view(-1, 1)
    y_test_tensor  = torch.tensor(y_test.values, dtype=torch.float32).view(-1, 1)

    full_train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    val_size = int(val_split * len(full_train_dataset))
    train_size = len(full_train_dataset) - val_size
    train_dataset, val_dataset = random_split(full_train_dataset, [train_size, val_size])

    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader   = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)
    test_dataset = TensorDataset(X_test_tensor, y_test_tensor)
    test_loader  = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)
    return train_loader, val_loader, test_loader


#################################
# 3. Pretrained Embeddings: GloVe
#################################
def load_glove_embeddings(filepath):
    """
    Load GloVe embeddings from a text file.
    Returns a dictionary mapping words to their embedding vectors.
    """
    embeddings = {}
    with open(filepath, 'r', encoding='utf-8') as f:
        for line in f:
            values = line.split()
            word = values[0]
            vector = np.asarray(values[1:], dtype='float32')
            embeddings[word] = vector
    return embeddings

def build_embedding_matrix(vocab, glove_embeddings, embedding_dim=100):
    """
    Create an embedding matrix where each row corresponds to a token in the vocabulary.
    If a word is not in the GloVe embeddings, initialize with a random normal vector.
    """
    embedding_matrix = np.zeros((len(vocab), embedding_dim))
    for word, idx in vocab.items():
        vector = glove_embeddings.get(word.lower())
        if vector is not None:
            embedding_matrix[idx] = vector
        else:
            embedding_matrix[idx] = np.random.normal(scale=0.6, size=(embedding_dim,))
    return torch.tensor(embedding_matrix, dtype=torch.float32)


#################################
# 4. Model Definitions
#################################
# --- Baseline Model: Randomly Initialized Embeddings ---
class LSTMSarcasmDetectorRandom(nn.Module):
    def __init__(self, vocab_size, embedding_dim, hidden_dim, n_layers, dropout):
        super(LSTMSarcasmDetectorRandom, self).__init__()
        # Embedding layer (randomly initialized)
        self.embedding = nn.Embedding(vocab_size, embedding_dim, padding_idx=0)
        # LSTM layer
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers=n_layers,
                            batch_first=True, dropout=dropout)
        # Fully connected layer for classification
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, text):
        embedded = self.embedding(text)
        output, (hidden, cell) = self.lstm(embedded)
        # Use the last hidden state
        hidden_last = hidden[-1]
        out = self.fc(hidden_last)
        return self.sigmoid(out)

# --- Improved Model: Pretrained GloVe + Dropout ---
class LSTMSarcasmDetectorPretrained(nn.Module):
    def __init__(self, vocab_size, embedding_tensor, hidden_dim, n_layers, dropout):
        super(LSTMSarcasmDetectorPretrained, self).__init__()
        # Load pretrained embeddings and allow fine-tuning (freeze=False)
        self.embedding = nn.Embedding.from_pretrained(embedding_tensor, freeze=False, padding_idx=0)
        # LSTM layer using the embedding dimension from the pretrained matrix
        self.lstm = nn.LSTM(embedding_tensor.shape[1], hidden_dim, num_layers=n_layers,
                            batch_first=True, dropout=dropout)
        # Fully connected classification layer
        self.fc = nn.Linear(hidden_dim, 1)
        self.sigmoid = nn.Sigmoid()

    def forward(self, text):
        embedded = self.embedding(text)
        output, (hidden, cell) = self.lstm(embedded)
        hidden_last = hidden[-1]
        out = self.fc(hidden_last)
        return self.sigmoid(out)


#################################
# 5. Training, Evaluation & Analysis Functions
#################################
# --- Training Without Early Stopping (Baseline Model) ---
def train_model_no_es(model, optimizer, criterion, train_loader, num_epochs, device):
    """Train the model without early stopping."""
    model.train()
    for epoch in range(num_epochs):
        running_loss = 0.0
        for batch_X, batch_y in tqdm(train_loader, desc=f"Baseline Epoch {epoch+1}/{num_epochs}"):
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        avg_loss = running_loss / len(train_loader)
        print(f"Baseline Epoch {epoch+1}/{num_epochs}, Loss: {avg_loss:.4f}")
    return model

# --- Training With Early Stopping (Improved Model) ---
def train_model_with_es(model, optimizer, criterion, train_loader, val_loader, device, num_epochs, patience):
    """
    Train the model with early stopping.
    If validation loss doesn't improve for 'patience' consecutive epochs, stop training.
    """
    best_val_loss = float('inf')
    trigger_times = 0
    best_model_state = None
    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        for batch_X, batch_y in tqdm(train_loader, desc=f"Improved Epoch {epoch+1}/{num_epochs}"):
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            optimizer.zero_grad()
            outputs = model(batch_X)
            loss = criterion(outputs, batch_y)
            loss.backward()
            optimizer.step()
            running_loss += loss.item()
        avg_train_loss = running_loss / len(train_loader)

        # Evaluate on validation set
        model.eval()
        val_loss = 0.0
        with torch.no_grad():
            for batch_X, batch_y in val_loader:
                batch_X, batch_y = batch_X.to(device), batch_y.to(device)
                outputs = model(batch_X)
                loss = criterion(outputs, batch_y)
                val_loss += loss.item()
        avg_val_loss = val_loss / len(val_loader)
        print(f"Improved Epoch {epoch+1}/{num_epochs} - Train Loss: {avg_train_loss:.4f}, Val Loss: {avg_val_loss:.4f}")

        if avg_val_loss < best_val_loss:
            best_val_loss = avg_val_loss
            best_model_state = deepcopy(model.state_dict())
            trigger_times = 0
        else:
            trigger_times += 1
            if trigger_times >= patience:
                print("Early stopping triggered for improved model.")
                break
    if best_model_state is not None:
        model.load_state_dict(best_model_state)
    return model

# --- Model Evaluation ---
def evaluate_model(model, loader, device):
    """Evaluate the model on a given DataLoader and return predictions and true labels."""
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for batch_X, batch_y in loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            outputs = model(batch_X)
            predictions = (outputs >= 0.5).float()
            all_preds.extend(predictions.cpu().numpy())
            all_labels.extend(batch_y.cpu().numpy())
    return np.array(all_preds).flatten(), np.array(all_labels).flatten()

# --- Misclassification Analysis ---
def get_misclassified_samples(model, test_loader, texts, vocab, device):
    """
    Identify false positives and false negatives from test predictions.
    Returns two lists containing the misclassified headlines.
    """
    model.eval()
    false_positives, false_negatives = [], []
    all_preds, all_labels = [], []
    with torch.no_grad():
        for batch_X, batch_y in test_loader:
            batch_X, batch_y = batch_X.to(device), batch_y.to(device)
            outputs = model(batch_X)
            predictions = (outputs >= 0.5).float()
            all_preds.extend(predictions.cpu().numpy().flatten())
            all_labels.extend(batch_y.cpu().numpy().flatten())

    all_preds = np.array(all_preds)
    all_labels = np.array(all_labels)
    for i, (pred, actual) in enumerate(zip(all_preds, all_labels)):
        if pred == 1 and actual == 0:
            false_positives.append(texts.iloc[i])
        elif pred == 0 and actual == 1:
            false_negatives.append(texts.iloc[i])
    return false_positives, false_negatives

# --- Word Importance Analysis ---
def compute_word_importance(model, text, vocab, device, max_length=20):
    """
    Compute word importance by calculating gradients of the model output
    with respect to the embeddings. Returns a list of tuples (word, importance score).
    """
    model.train()  # Temporarily set to training mode to enable backpropagation
    tokenized = torch.tensor([pad_sequence(text_to_sequence(text, vocab), max_length)], dtype=torch.long, device=device)
    # Forward pass through embedding layer
    embedded = model.embedding(tokenized)
    embedded.retain_grad()  # Retain gradients for the embeddings

    # Forward pass through LSTM and FC layer
    output, _ = model.lstm(embedded)
    hidden_last = output[:, -1, :]
    logits = model.fc(hidden_last)
    logits.backward()

    # Sum absolute gradients across the embedding dimensions for each word
    grads = embedded.grad.abs().sum(dim=-1).squeeze().cpu().numpy()
    if grads.max() != 0:
        grads = grads / grads.max()
    words = text.split()
    word_importance = list(zip(words[:max_length], grads[:len(words)]))
    model.eval()  # Switch back to evaluation mode
    return word_importance

def aggregate_word_importance(headlines, model, vocab, device, max_length=20):
    """
    Compute and aggregate word importance scores for a list of headlines.
    Returns a dictionary mapping each word to its average importance.
    """
    importance_totals = {}
    counts = {}
    for text in headlines:
        word_imp = compute_word_importance(model, text, vocab, device, max_length)
        for word, score in word_imp:
            importance_totals[word] = importance_totals.get(word, 0) + score
            counts[word] = counts.get(word, 0) + 1
    avg_importance = {word: importance_totals[word] / counts[word] for word in importance_totals}
    return avg_importance

def display_top_words(avg_importance, top_n=10):
    """Display the top N words based on average importance."""
    sorted_words = sorted(avg_importance.items(), key=lambda x: x[1], reverse=True)
    print(f"Top {top_n} words by average importance:")
    for word, score in sorted_words[:top_n]:
        print(f"{word}: {score:.4f}")


#################################
# 6. Main Routine: Data, Training & Evaluation
#################################
def main():
    # -------------------------------
    # Data Loading & Splitting
    # -------------------------------
    texts, labels = load_data('Sarcasm_Headlines_Dataset.json')
    X_train, X_test, y_train, y_test = split_data(texts, labels)

    # -------------------------------
    # Build Vocabulary and Preprocess Texts
    # -------------------------------
    vocab_local = build_vocab(X_train)
    max_length = 20  # Maximum sequence length
    X_train_seq = preprocess_texts(X_train, vocab_local, max_length)
    X_test_seq  = preprocess_texts(X_test, vocab_local, max_length)

    # -------------------------------
    # Create DataLoaders for Training, Validation, and Testing
    # -------------------------------
    train_loader, val_loader, test_loader = create_dataloaders(X_train_seq, X_test_seq, y_train, y_test)

    # -------------------------------
    # Load Pretrained GloVe Embeddings & Build Embedding Matrix
    # -------------------------------
    glove_filepath = "glove.6B.100d.txt"
    glove_embeddings = load_glove_embeddings(glove_filepath)
    embedding_dim = 100  # Using 100-dimensional embeddings
    embedding_tensor_local = build_embedding_matrix(vocab_local, glove_embeddings, embedding_dim)

    # -------------------------------
    # Set device (GPU if available)
    # -------------------------------
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # -------------------------------
    # Train Baseline Model (Random Embeddings)
    # -------------------------------
    baseline_model = LSTMSarcasmDetectorRandom(len(vocab_local), embedding_dim, hidden_dim=128, n_layers=2, dropout=0.0)
    baseline_model.to(device)
    criterion = nn.BCELoss()
    baseline_optimizer = optim.Adam(baseline_model.parameters(), lr=0.001, weight_decay=0.0)

    print("\nTraining Baseline Model (Random Embeddings):")
    baseline_model = train_model_no_es(baseline_model, baseline_optimizer, criterion, train_loader, num_epochs=10, device=device)

    # Evaluate Baseline Model on Test and Training Sets
    preds_baseline, true_labels_baseline = evaluate_model(baseline_model, test_loader, device)
    baseline_test_acc = accuracy_score(true_labels_baseline, preds_baseline)
    print(f"Baseline Model Test Accuracy: {baseline_test_acc * 100:.2f}%")

    preds_baseline_train, true_labels_baseline_train = evaluate_model(baseline_model, train_loader, device)
    baseline_train_acc = accuracy_score(true_labels_baseline_train, preds_baseline_train)
    print(f"Baseline Model Training Accuracy: {baseline_train_acc * 100:.2f}%")

    # Additional Metrics and Confusion Matrix for Baseline Model
    baseline_precision = precision_score(true_labels_baseline, preds_baseline)
    baseline_recall = recall_score(true_labels_baseline, preds_baseline)
    baseline_f1 = f1_score(true_labels_baseline, preds_baseline)
    print(f"Baseline Model Precision: {baseline_precision:.4f}")
    print(f"Baseline Model Recall: {baseline_recall:.4f}")
    print(f"Baseline Model F1-Score: {baseline_f1:.4f}")

    baseline_cm = confusion_matrix(true_labels_baseline, preds_baseline)
    plt.figure(figsize=(6, 4))
    sns.heatmap(baseline_cm, annot=True, fmt="d", cmap="Blues")
    plt.title("Baseline Model Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    false_positives_baseline, false_negatives_baseline = get_misclassified_samples(baseline_model, test_loader, X_test, vocab_local, device)
    print("\nBaseline Model False Positives (first 5):")
    print(false_positives_baseline[:5])
    print("\nBaseline Model False Negatives (first 5):")
    print(false_negatives_baseline[:5])

    # -------------------------------
    # Train Improved Model (Pretrained GloVe, Dropout, Early Stopping)
    # -------------------------------
    improved_model = LSTMSarcasmDetectorPretrained(len(vocab_local), embedding_tensor_local, hidden_dim=128, n_layers=2, dropout=0.1)
    improved_model.to(device)
    improved_optimizer = optim.Adam(improved_model.parameters(), lr=0.001, weight_decay=1e-7)

    print("\nTraining Improved Model (Pretrained GloVe, Dropout, Early Stopping):")
    improved_model = train_model_with_es(improved_model, improved_optimizer, criterion, train_loader, val_loader, device, num_epochs=10, patience=3)

    # Evaluate Improved Model on Test and Training Sets
    preds_improved, true_labels = evaluate_model(improved_model, test_loader, device)
    improved_test_acc = accuracy_score(true_labels, preds_improved)
    print(f"Improved Model Test Accuracy: {improved_test_acc * 100:.2f}%")

    preds_improved_train, true_labels_improved_train = evaluate_model(improved_model, train_loader, device)
    improved_train_acc = accuracy_score(true_labels_improved_train, preds_improved_train)
    print(f"Improved Model Training Accuracy: {improved_train_acc * 100:.2f}%")

    improved_precision = precision_score(true_labels, preds_improved)
    improved_recall = recall_score(true_labels, preds_improved)
    improved_f1 = f1_score(true_labels, preds_improved)
    print(f"Improved Model Precision: {improved_precision:.4f}")
    print(f"Improved Model Recall: {improved_recall:.4f}")
    print(f"Improved Model F1-Score: {improved_f1:.4f}")

    improved_cm = confusion_matrix(true_labels, preds_improved)
    plt.figure(figsize=(6, 4))
    sns.heatmap(improved_cm, annot=True, fmt="d", cmap="Blues")
    plt.title("Improved Model Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    false_positives_improved, false_negatives_improved = get_misclassified_samples(improved_model, test_loader, X_test, vocab_local, device)
    print("\nImproved Model False Positives (first 5):")
    print(false_positives_improved[:5])
    print("\nImproved Model False Negatives (first 5):")
    print(false_negatives_improved[:5])

    # -------------------------------
    # Word Importance Visualization on a Sample Headline
    # -------------------------------
    sample_text = X_test.iloc[0]
    word_importance = compute_word_importance(improved_model, sample_text, vocab_local, device, max_length)
    print("\nWord Importance for a Sample Headline:")
    for word, score in word_importance:
        print(f"{word}: {score:.4f}")

    # -------------------------------
    # Print Examples with Word Importance Scores
    # -------------------------------
    def print_headline_word_scores(headlines):
        for headline in headlines:
            print("Headline:", headline)
            scores = compute_word_importance(improved_model, headline, vocab_local, device, max_length)
            for word, score in scores:
                print(f"  {word}: {score:.4f}")
            print("\n")

    sarcastic_headlines = X_test[y_test == 1].iloc[:3]
    non_sarcastic_headlines = X_test[y_test == 0].iloc[:3]

    print("=== 3 Sarcastic Headlines and Word Importance ===")
    print_headline_word_scores(sarcastic_headlines)

    print("=== 3 Non-Sarcastic Headlines and Word Importance ===")
    print_headline_word_scores(non_sarcastic_headlines)

    print("=== 3 False Positives and Word Importance ===")
    print_headline_word_scores(false_positives_improved[:3])

    print("=== 3 False Negatives and Word Importance ===")
    print_headline_word_scores(false_negatives_improved[:3])

    # Return key variables for further analysis if needed
    return {
        "improved_model": improved_model,
        "X_test": X_test,
        "vocab_local": vocab_local,
        "test_loader": test_loader,
        "device": device,
        "y_test": y_test
    }

# Run main routine and capture returned objects
results = main()
improved_model = results["improved_model"]
vocab_local = results["vocab_local"]
device = results["device"]


#################################
# 7. Interactive GUI for Prediction & Visualization
#################################
def predict_and_visualize(headline):
    """
    Given a headline, preprocess it, obtain the prediction probability,
    compute word importance, and return an HTML visualization.
    """
    max_length = 20  # Must match training settings
    seq = pad_sequence(text_to_sequence(headline, vocab_local), max_length)
    tensor_seq = torch.tensor([seq], dtype=torch.long, device=device)

    improved_model.eval()
    with torch.no_grad():
        prob = improved_model(tensor_seq).item()
    prediction = "Sarcastic" if prob >= 0.5 else "Not Sarcastic"

    # Compute word importance scores
    word_imp = compute_word_importance(improved_model, headline, vocab_local, device, max_length)

    # Color each word based on its importance (white: low, red: high)
    cmap = mcolors.LinearSegmentedColormap.from_list("importance", ["white", "red"])
    norm = mcolors.Normalize(vmin=0, vmax=1)
    colored_words = []
    for word, weight in word_imp:
        color = mcolors.to_hex(cmap(norm(weight)))
        colored_words.append(f'<span style="background-color:{color}; padding:2px; margin:2px; border-radius:3px;">{word}</span>')
    html_words = " ".join(colored_words)

    html_output = f"""
    <p><b>Headline:</b> {headline}</p>
    <p><b>Prediction:</b> {prediction} (Probability: {prob:.4f})</p>
    <p><b>Word Importance:</b><br>{html_words}</p>
    """
    return html_output

# Create widgets for interactive prediction
text_input = widgets.Text(
    value="",
    placeholder="Enter a headline here",
    description="Headline:",
    disabled=False
)

predict_button = widgets.Button(
    description="Predict and Visualize",
    disabled=False,
    button_style="info",
    tooltip="Click to predict and visualize word importance",
    icon="check"
)

output_area = widgets.Output()

def on_button_click(b):
    with output_area:
        output_area.clear_output()
        html_output = predict_and_visualize(text_input.value)
        display(HTML(html_output))

predict_button.on_click(on_button_click)
display(text_input, predict_button, output_area)
```
